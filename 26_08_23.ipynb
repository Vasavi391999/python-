```python
import numpy as np
x=np.arange(0,10)
y=np.arange(11,21)
x
```

```python
y
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
y=x*x
plt.plot(x,y,'b*')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'v--')
```

```python
labels='python','c++','ruby','java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.4,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=False)
plt.axis('equal')
plt.show()
```
